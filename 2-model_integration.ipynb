## Models Integration with OpenAI , Google Gemini and GROQ

```python
# model info
import os
from dotenv import load_dotenv
load_dotenv()

google_api_key = os.getenv("GOOGLE_API_KEY") or os.getenv("GOOGLE_GENAI_API_KEY")
if google_api_key:
    os.environ["GOOGLE_API_KEY"] = google_api_key

groq_api_key = os.getenv("GROQ_API_KEY")
if groq_api_key:
    os.environ["GROQ_API_KEY"] = groq_api_key
```

```python
from langchain.chat_models import init_chat_model
# choose a single model string (init_chat_model expects a string, not a list)
if groq_api_key:
    model = init_chat_model("llama-3.1-70b-versatile", model_provider="groq")
elif google_api_key:
    model = init_chat_model("gemini-1.5-pro", model_provider="google_genai")
else:
    raise RuntimeError("No API key available for Groq or Gemini")
model
```

```python
## invoke the model with simple prompt

if getattr(model, "model_name", "") == "llama-3.1-70b-versatile":
    model = init_chat_model("llama-3.3-70b-versatile", model_provider="groq")

response = model.invoke("Hello how are you?")
print(response)
```

## Streaming and Batch

### Streaming
Most model can stream their output content while it is being generated. By displaying output progressively, streaming significantly improves user experience particularly for longer reponses. Calling stream() returns an iterator an interator that yields output chunks as they are produced. You can use a loop to process each chunk in real-time.

```python
model.invoke("Write a 200 word essay on the history of the internet")
```

```python
for chunk in model.stream("Write a 200 word essay on the history of the internet"):
    print(chunk.content, end = "", flush=True)

# by using the stream method you can see the output will its generated instead of waiting for the whole response to be generated and then printing it.
```

## Batch
Batching a collection of independent request to a model can significantly improve performance and reduce costs, as the processing can be done in parallel.

```python
response = model.batch([
    "Why do parrots talk?",
    "How much water an elephant drink in a day?"
])
for resp in response:
    print(resp.content)
```
